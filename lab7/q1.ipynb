```python
#q1
import math

def print_board(b):
    for row in b:
        print(" | ".join(row))
        print("-"*9)

def is_winner(b, p):
    # rows, cols, diags
    for i in range(3):
        if all(b[i][j]==p for j in range(3)): return True
        if all(b[j][i]==p for j in range(3)): return True
    if b[0][0]==p and b[1][1]==p and b[2][2]==p: return True
    if b[0][2]==p and b[1][1]==p and b[2][0]==p: return True
    return False

def is_full(b):
    return all(b[i][j] != " " for i in range(3) for j in range(3))

def minimax(b, is_max):
    if is_winner(b, "X"): return  1
    if is_winner(b, "O"): return -1
    if is_full(b):         return  0

    if is_max:
        best = -math.inf
        for i in range(3):
            for j in range(3):
                if b[i][j]==" ":
                    b[i][j] = "X"
                    val = minimax(b, False)
                    b[i][j] = " "
                    best = max(best, val)
        return best
    else:
        best = math.inf
        for i in range(3):
            for j in range(3):
                if b[i][j]==" ":
                    b[i][j] = "O"
                    val = minimax(b, True)
                    b[i][j] = " "
                    best = min(best, val)
        return best

def find_best_move(b):
    best_val = -math.inf
    move = None
    for i in range(3):
        for j in range(3):
            if b[i][j]==" ":
                b[i][j] = "X"
                val = minimax(b, False)
                b[i][j] = " "
                if val > best_val:
                    best_val, move = val, (i, j)
    return move

def main():
    board = [[" "]*3 for _ in range(3)]
    while True:
        print_board(board)
        # Human turn
        while True:
            try:
                r, c = map(int, input("Enter row and column (0-2): ").split())
                if 0 <= r <= 2 and 0 <= c <= 2:  # Check if input is within the board boundaries
                    break  # Exit the inner loop if input is valid
                else:
                    print("Invalid input. Row and column must be between 0 and 2.")
            except ValueError:
                print("Invalid input. Please enter two numbers separated by a space.")
        if board[r][c] != " ":
            print("Invalidâ€”try again.")
            continue
        board[r][c] = "O"
        if is_winner(board, "O"):
            print_board(board)
            print("You win!")
            break
        if is_full(board):
            print_board(board)
            print("Draw!")
            break

        # AI turn
        i, j = find_best_move(board)
        board[i][j] = "X"
        if is_winner(board, "X"):
            print_board(board)
            print("AI wins!")
            break
        if is_full(board):
            print_board(board)
            print("Draw!")
            break

if __name__ == "__main__":
    main()
```
